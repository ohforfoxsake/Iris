```python
setosa = iris.loc[0:49]
versicolor = iris.loc[50:99]
virginica = iris.loc[100:149]

setosa = setosa.drop(columns = 'class')
versicolor = versicolor.drop(columns = 'class')
virginica = virginica.drop(columns = 'class')
```

```python
sp_mean = iris.groupby('class').mean()
to_mean = iris.mean()
means = sp_mean.append(to_mean, ignore_index=True)
means.insert(0, 'class', ['setosa', 'versicolor', 'virginica', 'all_species'])
means
```

```python
sp_std = iris.groupby('class').std()
to_std = iris.std()
std = sp_std.append(to_std, ignore_index=True)
std.insert(0, 'class', ['setosa', 'versicolor', 'virginica', 'all_species'])
std
```
